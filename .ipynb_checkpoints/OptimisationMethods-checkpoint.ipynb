```python
#Import stuff
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import math
import sklearn
import sklearn.datasets

from opt_utils import load_params_and_grads
from opt_utils import plot_decision_boundary, load_dataset
from testCases import *

%matplotlib inline
plt.rcParams['figure.figsize'] = (7.0, 4.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
def model(X, Y, layer_dims, optimizer, learning_rate=0.0007, num_epochs= 1000, mini_batch_size=64, beta1=0.9,
          beta2=0.999, epsilon=1e-8, lambd=0.7, dropOutProb=0.7, print_cost=True):
    
    costs = []
    t = 0
    
    #Initialize Parameters
    parameters = initializeParameters(layer_dims)
    
    #Initialize Optimizer
    V, S = initializeOptimizer(parameters)
    
    #Optimization Loop
    for i in range(num_epochs):
        minibatches = random_mini_batches(X, Y, mini_batch_size)
        for minibatch in minibatches:
            (minibatch_X, minibatch_Y) = minibatch
            
            #Forward Propagation
            AL, caches = forwardPropagation(minibatch_X, parameters, dropOutProb=0.5)
            
            #Calculate Cost
            cost = calculateCost(AL, minibatch_Y, parameters, lambd)
                
            #Back Propagation
            grads = backpropagation(AL, minibatch_Y, caches, learning_rate, lambd, dropOutProb, epsilon)
            
            #Update Parameters
            t = t + 1
            parameters, V, S = updateParameters(optimizer, grads, parameters, learning_rate, V, S, beta1, beta2, t, epsilon)
            
        if print_cost and i%1000==0:
            print("Cost at iteration " + str(i) + " is " + str(cost))
        if i%100==0:
            costs.append(cost)
    
    # plot the cost
    plt.plot(costs)
    plt.ylabel('cost')
    plt.xlabel('epochs (per 100)')
    plt.title("Learning rate = " + str(learning_rate))
    plt.show()
    
    return parameters

def predict(X, Y, parameters):
    m = X.shape[1]
    p = np.zeros((1, m), dtype=float)
    A, caches = forwardPropagation(X, parameters, 0.5)
    for i in range(m):
        if A[0, i] >= 0.5:
            p[0, i] = 1
        else:
            p[0, i] = 0
    print("Accracy: " + str(np.mean((p[0,:] == Y[0,:]))))
    return p
```

# Helper Functions

```python
def initializeParameters(layer_dims):
    parameters = {}
    L = len(layer_dims)
    for l in range(1, L):
        parameters["W" + str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) * np.sqrt(2/layer_dims[l-1])
        parameters["b" + str(l)] = np.zeros((layer_dims[l], 1), dtype=float)
    return parameters
```

```python
def initializeOptimizer(parameters):
    L = len(parameters)//2
    V = {}
    S = {}
    for l in range(L):
        r = parameters["W" + str(l+1)].shape[0]
        c = parameters["W" + str(l+1)].shape[1]
        V["dW" + str(l+1)] = np.zeros((r, c), dtype=float)
        V["db" + str(l+1)] = np.zeros((r, 1), dtype=float)
        S["dW" + str(l+1)] = np.zeros((r, c), dtype=float)
        S["db" + str(l+1)] = np.zeros((r, 1), dtype=float)
    return V, S
```

```python
def random_mini_batches(X, Y, mini_batch_size):
    m = X.shape[1]
    minibatches = []
    
    #STEP1: Shuffle the data
    permutation = list(np.random.permutation(m))
    shuffled_X = X[:, permutation]
    shuffled_Y = Y[:, permutation]
    
    #STEP2: Create Partition
    num_complete_batches = math.floor(m/mini_batch_size)
    for k in range(num_complete_batches):
        mini_batch_X = shuffled_X[:, k*mini_batch_size : (k+1)*mini_batch_size]
        mini_batch_Y = shuffled_Y[:, k*mini_batch_size : (k+1)*mini_batch_size]
        mini_batch = (mini_batch_X, mini_batch_Y)
        minibatches.append(mini_batch)
    
    if m%mini_batch_size != 0:
        mini_batch_X = shuffled_X[:, num_complete_batches * mini_batch_size :]
        mini_batch_Y = shuffled_Y[:, num_complete_batches * mini_batch_size :]
        mini_batch = (mini_batch_X, mini_batch_Y)
        minibatches.append(mini_batch)
    
    return minibatches
```

```python
#Forward Propagation

def forwardPropagation(X, parameters, dropOutProb):
    m = X.shape[1]
    A = X
    L = len(parameters)//2
    caches = []
    for l in range(1, L):
        A_prev = A
        W = parameters["W" + str(l)]
        b = parameters["b" + str(l)]
        A, cache = linearActivationForward(A_prev, W, b, "relu", dropOutProb)
        caches.append(cache)
    A_prev = A
    W = parameters["W" + str(L)]
    b = parameters["b" + str(L)]
    AL, cache = linearActivationForward(A_prev, W, b, "sigmoid", dropOutProb)
    caches.append(cache)
    return AL, caches
    

def linearActivationForward(A_prev, W, b, activation, dropOutProb):
    Z, linearCache = linearForward(A_prev, W, b)
    if activation == "relu":
        A_pre, activationCache = relu(Z)
    else:
        A_pre, activationCache = sigmoid(Z)
    A, dropOutCache = dropOut(A_pre, dropOutProb)
    cache = (linearCache, activationCache, dropOutCache)
    return A, cache

def linearForward(A_prev, W, b):
    Z = np.dot(W, A_prev) + b
    cache = (A_prev, W, b)
    return Z, cache

def relu(Z):
    A = np.maximum(0, Z)
    cache = Z
    return A, cache

def sigmoid(Z):
    A = 1/(1 + np.exp(-Z))
    cache = Z
    return A, cache

def dropOut(A_pre, dropOutProb):
    D = np.random.randn(A_pre.shape[0], A_pre.shape[1])
    D = D < dropOutProb
    A = np.multiply(A_pre, D)
    A = A/dropOutProb
    cache = D
    return A, cache
```

```python
#Calculate Cost

def calculateCost(A, Y, parameters, lambd):
    m = Y.shape[1]
    L = len(parameters)//2
    fro_norm = 0.0
    for l in range(L):
        fro_norm += np.sum(np.square(parameters["W" + str(l+1)]))

    normal_cost = np.dot(Y, np.log(A.T)) + np.dot(1-Y, np.log(1-A.T))
#     normal_cost = np.sum(np.multiply(-np.log(A),Y) + np.multiply(-np.log(1 - A), 1 - Y))
    
    print("normal_cost: " + str(normal_cost))
    print("-------------------")
    
    cost = (1./m) * normal_cost + (lambd/(2*m)) * fro_norm
    cost = np.squeeze(cost)
    return cost
```

```python
#Backpropagation

def backpropagation(AL, Y, caches, learning_rate, lambd, dropOutProb, epsilon):
    grads = {}
    L = len(caches)
    m = Y.shape[1]
    current_cache = caches[L-1]
    grads["dA" + str(L)] = - (np.divide(Y, AL + epsilon) - np.divide(1-Y, (1-AL)+epsilon))
#   grads["dA" + str(L)] = - (np.divide(Y, AL)- np.divide(1-Y, 1-AL))
    grads["dA" + str(L)] = dropOutBackwards(grads["dA" + str(L)], dropOutProb, current_cache[2])
    grads["dA" + str(L-1)], grads["dW" + str(L)], grads["db" + str(L)] = linearActivationBackwards(grads["dA" + str(L)], current_cache, lambd, "sigmoid")
    for l in reversed(range(L-1)):
        current_cache = caches[l]
        grads["dA" + str(l)], grads["dW" + str(l+1)], grads["db" + str(l+1)] = linearActivationBackwards(grads["dA" + str(l+1)], current_cache, lambd, "sigmoid")
        if l > 0:
            grads["dA" + str(l)] = dropOutBackwards(grads["dA" + str(l)], dropOutProb, caches[l-1][2])
    return grads
    
def dropOutBackwards(dA, dropOutProb, dropOutCache):
    D = dropOutCache
    dA = np.multiply(dA, D)
    dA = dA/dropOutProb
    return dA

def linearActivationBackwards(dA, cache, lambd, activation):
    (linearCache, activationCache, dropOutCache) = cache
    if activation == "sigmoid":
        dZ = sigmoidBackwards(dA, activationCache)
    else:
        dZ = reluBackwards(dA, activationCache)
    dA_prev, dW, db = linearBackwards(dZ, linearCache, lambd)
    return dA_prev, dW, db

def sigmoidBackwards(dA, cache):
    Z = cache
    s = 1 / (1 + np.exp(-Z))
    dZ = dA * s * (1 - s)
    return dZ

def reluBackwards(dA, cache):
    Z = cache
    dZ = np.array(dA, copy=True)
    dZ[Z<=0] = 0
    return dZ

def linearBackwards(dZ, cache, lambd):
    (A_prev, W, b) = cache
    m = A_prev.shape[1]
    dW = (1/m) * np.dot(dZ, A_prev.T) + (lambd/m) * W
    db = (1/m) * np.sum(dZ, axis=1, keepdims=True)
    dA_prev = np.dot(W.T, dZ)
    return dA_prev, dW, db
```

```python
#Update Parameters

def updateParameters(optimizer, grads, parameters, learning_rate, V, S, beta1, beta2, t, epsilon):
    
    L = len(parameters)//2
    for l in range(L):
        V["dW" + str(l+1)] = (beta1 * V["dW" + str(l+1)] + (1 - beta1) * grads["dW" + str(l+1)])/(1 - pow(beta1, t))
        V["db" + str(l+1)] = (beta1 * V["db" + str(l+1)] + (1 - beta1) * grads["db" + str(l+1)])/(1 - pow(beta1, t))
        
        if optimizer == "momentum":
            parameters["W" + str(l+1)] -= learning_rate * V["dW" + str(l+1)]
            parameters["b" + str(l+1)] -= learning_rate * V["db" + str(l+1)]
        elif optimizer == "adam":
            S["dW" + str(l+1)] = (beta2 * S["dW" + str(l+1)] + (1 - beta2) * np.power(grads["dW" + str(l+1)], 2))/(1 - pow(beta2, t))
            S["db" + str(l+1)] = (beta2 * S["db" + str(l+1)] + (1 - beta2) * np.power(grads["db" + str(l+1)], 2))/(1 - pow(beta2, t))
            parameters["W" + str(l+1)] -= learning_rate * np.divide(V["dW" + str(l+1)], np.sqrt(S["dW" + str(l+1)]) + epsilon)
            parameters["b" + str(l+1)] -= learning_rate * np.divide(V["db" + str(l+1)], np.sqrt(S["db" + str(l+1)]) + epsilon)
    return parameters, V, S
```
